```python
%matplotlib inline

import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def one_iterate(pile):
    give_poses = np.where(pile[1:-1,1:-1] > 1.)
    poses = zip(give_poses[0],give_poses[1])
    for pos in poses:
        i = pos[0] + 1
        j = pos[1] + 1
        gain = pile[i, j]/4.
        pile[i-1, j] += gain
        pile[i+1, j] += gain
        pile[i, j-1] += gain
        pile[i, j+1] += gain
        pile[i,j] = 0
    return pile

def one_avalanche(old_pile, N = 10):
    i,j = np.random.randint(1, N+1, size=2)
    pile = old_pile
    pile[i,j] += np.random.random()
    while np.sum(np.where(pile > 1, 1, 0)) >= 1:
        pile = one_iterate(pile)
    avalanche_size = np.sum(np.where(abs(pile - old_pile) > 0.01, 1, 0))
    return pile, avalanche_size

def iterate(N = 10, tmax = 80):
    pile = np.zeros([N+2, N+2], dtype = np.float32)
    for t in range(tmax):
        fig, ax = plt.subplots(ncols = 1)
        pile, avalanche_size = one_avalanche(pile)
        ax.matshow(pile)
        plt.show()

iterate()
```
